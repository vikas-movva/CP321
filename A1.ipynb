```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
# change date to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
# Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.
df_2023 = df[df['Date'].dt.year == 2023]
df_2023.head()
```

```python
# filter for fuel type 'Regular Unleaded Gasoline', select only the columns with numerical values and drop the '_id' column
df_2023_regular = df_2023[df_2023['Fuel Type'] == 'Regular Unleaded Gasoline'].select_dtypes(include=['float64', 'int64']).drop(columns=['_id', 'Ontario Average/Moyenne provinciale', 'Southern Average/Moyenne du sud de l\'Ontario', 'Northern Average/Moyenne du nord de l\'Ontario'])
mean_values = df_2023_regular.mean()
mean_values
```

```python
# plot the bar chart sorted by the price amounts
mean_values.sort_values().plot(kind='barh')
```

## Question 2
Plot the corresponding dot plot for the same information presented in Question 1.

```python
# Plot the corresponding dot plot for the same information presented in Question 1.
mean_values = mean_values.sort_values() #.plot(kind='scatter', x=mean_values.index, y=mean_values.values)
#swap x-axis for y-axis and y-axis for x-axis
fig, ax = plt.subplots()
ax.scatter(x=mean_values.values, y=mean_values.index)
# plt.xticks(range(len(mean_values)), mean_values.index, rotation='vertical')
plt.ylabel('City')
plt.xlabel('Price')
plt.show()
```

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
# Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins. 
df_2023_regular['Toronto West/Ouest'].hist(bins=15)
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
# Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.
# sort the prices in ascending order
sorted_prices = np.sort(df_2023_regular['Toronto West/Ouest'])
# calculate the y-axis values for the ECDF
y_values = np.arange(1, len(sorted_prices) + 1) / len(sorted_prices)
# plot the ECDF
plt.plot(sorted_prices, y_values, marker='.', linestyle='solid')
plt.xlabel('Regular Unleaded Gasoline Price')
plt.ylabel('ECDF')
plt.show()
```
