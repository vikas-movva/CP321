```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('data.csv')
```

```python
df['Wages'].unique()
```

```python
df_filtered = df[(df['Geography'] != 'Canada') & (df['Type of work'] != 'Both full- and part-time') & (df['Education level'] != 'PSE  (5,6,7,8,9))') & (df['Education level'] != 'No PSE  (0,1,2,3,4)') & (df['Education level'] != '   University degree') & (df['Education level'] != 'Total, all education levels') & (df['Wages'] == 'Average hourly wage rate')]
df_filtered.sort_values(by='Both Sexes', ascending=False)
```

```python
df_fulltime = df[df['Type of work'] == '   Full-time ']
df_fulltime
```

```python
average = df[['Male', 'Female']].mean()
average.plot(kind='bar')
plt.xlabel('Sex')
plt.ylabel(' Average wage')
plt.title('Average wage by Sex')
plt.show()
```

```python
# 2. Wages by education level
df['Education level'].unique()
```

```python
# Average wage by education level
df.groupby('Education level')['Both Sexes'].mean().sort_values().plot(kind='bar')
plt.xlabel('Education level')
plt.ylabel('Average wage')
plt.title('Average wage by education level')
```

```python
# Wages by age group
df.groupby('Age group')['Both Sexes'].mean().sort_values().plot(kind='bar')
plt.xlabel('Age group')
plt.ylabel('Average wage')
plt.title('Average wage by age group')
```

```python
# Wages by type of work
df.groupby('Type of work')['Both Sexes'].mean().sort_values().plot(kind='bar')
plt.xlabel('Type of work')
plt.ylabel('Average wage')
plt.title('Average wage by type of work')
```

```python
# Wages by type of location
df.groupby('Geography')['Both Sexes'].mean().sort_values().plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Average wage')
plt.title('Average wage by Geography')
```

```python
df.groupby('YEAR')['Both Sexes'].mean().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Average wage')
plt.title('Average wage by year')
```

```python
df.groupby('YEAR')['Both Sexes'].mean().plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average wage')
plt.title('Average wage by year')
```
